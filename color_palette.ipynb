```
import os
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
from sklearn.cluster import KMeans
```

```
# Upload any image to Google colab to test.
os.listdir()
```

```
# Let's use an uploaded image named sample.jpg
# Feel free to change the filename
filename = 'sample.jpg'
img = Image.open(filename).convert('RGB')
```

```
# Let's view the image.
plt.imshow(img)
plt.axis('off')
```

```
# The image is now represented as a 3D matrix. Two dimensions to represent 
#the location of the pixel and the third dimension  to contain the RGB values.
arr = np.array(img)
```

```
# The shape represents (height, width, 3).
# Notice how the last value is explicitly 3. One for each in RGB.
arr.shape
```

```
# We want the a 2D array for each pixel and their RGB values.
# Notice how we specified the dimensions to be a 2D array keeping the RGB values for each pixel.
arr = np.reshape(arr, (-1, 3))
```

```
# Sanity check: Make sure the shape of the array is (height x width, 3)
arr.shape
```

```
# We want 6 colors so let's specify our clusters to 6.
kmeans = KMeans(n_clusters=6, random_state=42).fit(arr)
color_palette = kmeans.cluster_centers_
```

```
# We'll need to round off the values for validity.
color_palette = np.around(color_palette).astype(int)
```

```
color_palette
```

```
# Let's plot our color_palette!
plt.imshow([color_palette])
plt.axis('off')
```
